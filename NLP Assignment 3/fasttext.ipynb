## FastText

```
import fasttext
```

### Training Model

```
params = {
    "epoch": 1,
    "dim": 30,
    "lr": 0.2,
    "wordNgrams": 1
}
model = fasttext.train_supervised(input = "train_fast.txt",**params)
```

### Reading Test Data

```
test_data = []
test_real = []
with open("test_fast.txt") as f:
  l = f.readlines()
  for i in l:
    ll = i.strip().split()
    test_data.append(' '.join(ll[1:]))
    test_real.append(ll[0])
```

```
test_data[:10]
```

### Running Inference

```
test_pred, _ = model.predict(test_data)
```

### Generating Metrics

```
from sklearn.metrics import classification_report
```

```
print(classification_report(test_real,test_pred))
```
